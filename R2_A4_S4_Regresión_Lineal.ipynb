```python
#Import the librart numpy to use the arrays and the matriz
import numpy as np
#Import the model to train
from sklearn.model_selection import train_test_split
#Import the LinearRegression
from sklearn.linear_model import LinearRegression
#Import the elements for creation of the table
import matplotlib.pyplot as plt

#Get the data for the analysis
Tourism_Expenditure = np.array([3,10,5,15,7,20,12,25,8,30,18,22,6,35,16,40,13,27,9,14,19,24,11,33])
Number_Tourists = np.array([550,700,600,800,650,900,750,1000,680,1100,850,950,620,1200,800,1300,700,1050,670,720,820,900,710,1150])

#With the function reshape change the form of the array and asigned the data for the axes X and Y
x = Number_Tourists.reshape(-1, 1)
y = Tourism_Expenditure

#Split the data for train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#Create the model
model = LinearRegression()

#Train the model
model.fit(x_train, y_train)

#Predict the data
y_pred = model.predict(x_test)

#Calculate the R2 score
r2 = model.score(x_test, y_test)

#Get the coefficients and the intercept
coefficient = model.coef_[0]
intercept = model.intercept_

print("Linear Regression")
print("\nR2 score:", r2)
print("Coefficient:", coefficient)
print("Intercept:", intercept)

print("\nCurrent Expenditure of Tourist VS Predicted")
for actual, predicted in zip(y_test, y_pred):
    print(f"Current: {actual:.2f}, Predicted: {predicted:.2f}")

print("\n")

plt.scatter(x, y, color='blue', label='Current')
plt.plot(x_test, y_pred, color='red', label='Predicted')
plt.xlabel('Number of Tourists')
plt.ylabel('Tourism Expenditure')
plt.title('Linear Regression Model: Tourism Expenditure vs Number of Tourists')
plt.legend()
plt.show()
```
