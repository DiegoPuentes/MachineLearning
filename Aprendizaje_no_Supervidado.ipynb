```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

#Load the Iris dataset
iris = load_iris()
X = iris.data  # We use all characteristics (sepal length, sepal width, petal length, petal width).

# Configuring and training the K-Means model
# We chose 3 clusters as we know that there are 3 flower species
kmeans = KMeans(n_clusters=3, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Reduce dimensionality for visualisation
# We use PCA to reduce to 2 dimensions
pca = PCA(2)
X_pca = pca.fit_transform(X)

#Visualising the results
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = pca.transform(kmeans.cluster_centers_)  #We transform the cluster centres to the same dimensions.
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Clustering de K-Means')
plt.show()

```
